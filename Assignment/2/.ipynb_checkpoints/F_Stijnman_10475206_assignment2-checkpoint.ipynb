```python
import numpy as np
import matplotlib.pyplot as plt


def gillespie(S, I, N, beta, gamma):
    Transmission = beta * S * I / N
    Recovery = gamma * I
    R_total = Transmission + Recovery
    
    rand_1 = np.random.rand()
    rand_2 = np.random.rand()
    
    delta_t = -1 / R_total * np.log(rand_1)
    if rand_2 < (Transmission / R_total):
        I += 1
    else:
        I -= 1
    return I, delta_t

def discrete_event(S, I, N, beta, gamma, tot_time):
    I_data = [I]
    x = [0]
    while x[-1] < tot_time and I_data[-1] > 0:
        
        I, t = gillespie(S, I_data[-1], N, beta, gamma)

        I_data.append(I)
        x.append(x[-1] + t)
        
    return I_data, x

beta = 0.03
gamma = 1 / 100.0
I = 70.0
N = 100.0
S = N - I

I_data, x = discrete_event(S, I, N, beta, gamma, 5*365)

plt.plot(np.array(x)/365, I_data)
plt.xlabel("Time (years)")
plt.ylabel("Number of Infected (I)")
plt.show()
```
